```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars= pd.read_csv('mtcars.csv')
cars.head()
```

```python
cars.shape
```

```python
cars.describe()
```

```python
#checking nullvlaues
cars.isna().sum()
```

```python
#replacing null values in wt column with average weight
cars['wt'].fillna(3.2,inplace=True)
```

```python
#replacing null values in gear column withaverage weight
cars['gear'].value_counts()
cars['gear'].fillna (3,inplace = True)
```

```python
cars.isna().sum()
```

```python
##EXPLORATORY DATA ANALYSIS
```

```python
#find cars whose highway mileage is equalto or above 18
cars[cars['mpg']>=20]
```

```python
#find cars whose highway mileage is equalto or above 25
cars[cars['mpg']>=25]
```

```python
cars[cars['mpg']>=25][['model','mpg','cyl','am','gear']]
```

```python
#cars with 5 gears
cars[cars['gear']==5]
```

```python
#cars with 5 gears and 8 cylinders
cars[(cars['gear']==5)&(cars['cyl']==8)]
```

```python
#cars with automatic gears
cars[cars['am']==1]
```

```python
#find average mileage based on gear type
avg_auto_mpg=np.mean(cars[cars['am']==1]['mpg'])
print('automatic cars average mileage ',avg_auto_mpg)
avg_man_mpg=np.mean(cars[cars['am']==0]['mpg'])
print('manual cars average mileage ',avg_man_mpg)
```

```python
#above same car ussing group by
cars.groupby ('am')['mpg'].mean()
```

```python
cylinders = cars.cyl.value_counts()
cylinders
#dustribution of cars as per cylinders
plt.pie(cylinders.values,labels=['8 Cyl','4 Cyl','6 Cyl'],colors=['pink','skyblue','purple'],
        autopct='%0.1f%%',explode=[0.02,0.02,0.02])
plt.show()
```

```python
#distribution of cars as manualand automatic
gear_types = cars['am'].value_counts()
gear_types
plt.pie(gear_types.values, labels=['automatic','manual'],colors=['skyblue','lightgreen'],
        autopct='%1.1f%%', explode=[0.02,0])
plt.title('Gear type distribution')
plt.show()
```

```python
#no of gears comparison
gears = cars.gear.value_counts()
gears
```

```python
#bar graph
plt.bar(gears.index,gears.values,color = '#672524')
plt.xlabel('number of gears')
plt.ylabel('count of cars')
plt.show()
```
